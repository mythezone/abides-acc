```python
logging.addLevelName(32,"ABIDES")
```

```python
formatter = logging.Formatter('%(time)f: %(message)s')
```

```python
import sys
handler = logging.StreamHandler(sys.stdout)
```

```python
handler.setFormatter(formatter)
```

```python
handler.setLevel(30)
```

```python
logger = logging.getLogger("ABIDES")
```

```python
logger.addHandler(handler)
```
