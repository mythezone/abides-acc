```python
from model.timeseries.traditional_generator import discrete_mean_reverting_fundamental, ornstein_uhlenbeck_process
from gui.draw.multi_timeseries import plot_multi_timeseries, plot_x_y


mts = []

for i in range(1):
    x = discrete_mean_reverting_fundamental(
        r_bar=100,
        sigma_s=2,
        kappa=0.2*i,
        length=20
    )
    mts.append(x)


plot_multi_timeseries(
    mts,
    title="Mean Reverting Fundamental",
    xlabel="Time",
    ylabel="Value",
    legend=["Series {}".format(i) for i in range(len(mts))]
)

```

```python
def descretization(x):
    l = len(x)
    res = [0]
    for i in range(1, l):
        res.append((x[i] - x[i-1])/x[i-1])

    return res 

des_x = [descretization(mts[0])]
plot_multi_timeseries(
    des_x,
    title="Mean Reverting Fundamental",
    xlabel="Time",
    ylabel="Value",
    legend=["Series {}".format(i) for i in range(len(mts))]
)


```

```python
def plot_diff_data(x):
    """
    绘制标准化后金融差分数据的曲线图，
    纵坐标为百分比，范围为 [-10%, +10%]，共16格。

    参数:
        x (array-like): 差分并标准化后的时序数据
    """
    x = np.asarray(x)
    y_ticks = np.linspace(-0.10, 0.10, 17)  # 16格，17个刻度点

    plt.figure(figsize=(12, 4))
    plt.plot(x, label="Standardized Return")
    plt.ylim(-0.10, 0.10)
    plt.yticks(y_ticks)
    plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=1))
    plt.grid(True, which="both", axis="y", linestyle="--", linewidth=0.5)
    plt.title("Differenced & Standardized Financial Time Series")
    plt.xlabel("Time")
    plt.ylabel("Return (%)")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_diff_data(des_x)
```
